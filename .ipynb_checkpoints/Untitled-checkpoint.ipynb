```python
import pandas as pd
import numpy as np
import datetime as dt

parser1 = lambda date: pd.datetime.strptime(date, '%d/%m/%Y')
parser2 = lambda date: pd.datetime.strptime(date, '%Y-%m-%d')
client_train = pd.read_csv("steg/train/client_train.csv", parse_dates=[4], date_parser=parser1)
invoice_train = pd.read_csv("steg/train/invoice_train.csv", parse_dates=[1], date_parser=parser2)
```

```python
client_train['client_id'] = client_train['client_id'].transform(lambda x: x.split('_')[2])
client_train.head()
```

```python
invoice_train['client_id'] = invoice_train['client_id'].transform(lambda x: x.split('_')[2])
invoice_train.head()
```

```python
df_train = client_train.merge(invoice_train, how='inner')
```

```python
df_train = df_train.drop(columns = ['old_index','new_index'])
df_train.head()
```

```python
# creating bool series True for NaN values  
bool_series = pd.isnull(df_train['counter_type'])  
    
# filtering data  
# displaying data only with Gender = NaN  
df_train[bool_series].shape
```

```python
df_train['client_id'] = df_train['client_id'].astype(int)
df_train.dtypes
```

```python
df_train['counter_type'] = df_train['counter_type'].astype(str)
df_train.dtypes
```

```python
dummy = pd.get_dummies(df_train['counter_type'])
df_train_one_hot = dummy.join(df_train)
df_train_one_hot = df_train_one_hot.drop(columns=['counter_type'])
df_train_one_hot.head()
```

```python
#Dropping id cuz useless
df_train_one_hot = df_train_one_hot.drop(columns=['client_id'])
df_train_one_hot.head()
```

```python
df_train_one_hot['target'].value_counts()
```

```python
from sklearn.utils import resample

# Separate majority and minority classes
df_majority = df_train_one_hot[df_train_one_hot.target==0]
df_minority = df_train_one_hot[df_train_one_hot.target==1]
 
# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=400000,     # to match minority class
                                 random_state=123) # reproducible results
 
# Combine minority class with downsampled majority class
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
 
# Display new class counts
df_downsampled.target.value_counts()
```

```python
#df_downsampled['counter_number'] = df_downsampled['counter_number'] * df_downsampled['counter_coefficient']
#df_downsampled = df_downsampled.drop(columns=['counter_coefficient'])
df_downsampled.counter_statue.value_counts()
```

```python
X = df_downsampled[['disrict','region','tarif_type','counter_number','counter_code','reading_remarque','consommation_level_1','consommation_level_2','consommation_level_3','consommation_level_4','months_number',11,12,51,'ELEC','GAZ']]
Y = df_downsampled['target']
```

```python
Y.shape
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split

X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.15)

D_train = xgb.DMatrix(X_train, label=Y_train)
D_test = xgb.DMatrix(X_val, Y_val)

param = {
    'eta': 0.3, 
    'max_depth': 3,  
    'objective': 'multi:softprob',  
    'num_class': 2} 

steps = 20  # The number of training iterations

model = xgb.train(param, D_train, steps)

preds = model.predict(D_test)
best_preds = np.asarray([np.argmax(line) for line in preds])
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_val, best_preds)
accuracy
```
